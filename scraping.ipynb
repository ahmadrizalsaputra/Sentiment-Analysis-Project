## **Scraping Mobile Legends: Bang Bang Review from Google Play Store**

#### **1. Import Library**

```python
import pandas as pd  
import numpy as np 
from google_play_scraper import app, Sort, reviews
```

#### **2. App Information**

```python
app_info = app("com.mobile.legends")
print("App Information:\n", app_info)
```

#### **3. Scraping**

```python
result, continuation_token = reviews( 
    'com.mobile.legends',
    lang='id',
    country='id',
    sort=Sort.NEWEST,
    count=10000,
    filter_score_with=None
)

df = pd.DataFrame(np.array(result), columns=['review'])
df = df.join(pd.DataFrame(df.pop('review').tolist()))
df.head()
```

```python
df.info()
```

#### **4. Drop Feature**

```python
df_new = df.drop(columns = ['reviewId','userName','reviewCreatedVersion','replyContent','repliedAt',
                            'appVersion','userImage','thumbsUpCount','at','score'])
```

```python
df_new.insert(0, "ID", range(1, len(df_new) + 1))
```

```python
df_new.info()
```

#### **5. Export to .csv**

```python
df_new.to_csv("hasil_scraping.csv", index=False)
```
